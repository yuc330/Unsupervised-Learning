```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
import scipy.io
mat = scipy.io.loadmat('hw1p1.mat')
data1 = (mat['hw1p1data'])
```

```python
plt.plot(data1[:,0], data1[:,1])
plt.text(5, 60, 'plt.plot(data1[:,0], data1[:,1])')
plt.title('hw1p1data')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
mat = scipy.io.loadmat('hw1p2.mat')
data2 = (mat['hw1p2data'])
fig = plt.figure(figsize = (20,10))
ax = fig.gca(projection='3d')
X = data2[:, 0].reshape(101,101)
Y = data2[:, 1].reshape(101,101)
Z = data2[:, 2].reshape(101,101)
surf = ax.plot_surface(X, Y, Z, cmap=cm.ocean)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('hw1p2data')
plt.show()
```
